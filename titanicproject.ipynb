```python
#loading csv file 
import pandas as pd
import csv
df=pd.read_excel(r"C:\Users\Sahana S B\OneDrive\Documents\Titanic-Dataset.xlsx",sheet_name="Titanic-Dataset",header=0)
df.to_csv(r"C:\Users\Sahana S B\OneDrive\Documents\Titanic-Dataset.csv",index="False")
csv.reader(r"C:\Users\Sahana S B\OneDrive\Documents\Titanic-Dataset.csv")
print(df)
```

```python
#checking for empty values
print(df.isnull().sum())
```

```python
df.dropna(subset=['Embarked'],inplace=True)
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Cabin'].fillna('unknown',inplace=True)
print(df)
print(df.isnull().sum())
```

```python
df['Survived']=df['Survived'].astype('category')
df['Pclass']=df['Pclass'].astype('category')
print(df)
```

```python
df.drop_duplicates(inplace=True)
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(df['Age'],bins=30,kde=True)
plt.title("Distribution of age:")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
plt.figure(figsize=(10,6))
sns.histplot(df['Fare'],bins=30,kde=True)
plt.title("Distribution of fare:")
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.countplot(x='Survived',data=df)
plt.title('Count of survived')
plt.xlabel('survived')
plt.ylabel('count')
plt.show()
```
